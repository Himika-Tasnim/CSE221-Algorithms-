```python
import heapq

def adjacency_list(vertex,edge):
    adj_list=[]
    for i in range(vertex+1):
        adj_list.append([])
    for i in range(edge):
        x=list(map(int,inp.readline().split()))
        adj_list[x[0]].append((x[1],x[2]))
    return adj_list

def dijkstra(adj_list,src,heap,distance,visited):
    lst=adj_list[src]
    visited[src]=True
    for j in range(len(lst)):
        if visited[lst[j][0]]==False and distance[lst[j][0]]>(lst[j][1]+distance[src]):
            distance[lst[j][0]]=lst[j][1]+distance[src]
            heapq.heappush(heap,(distance[lst[j][0]],lst[j][0]))

    return distance  



inp=open("input2.txt","r")
out=open("output2.txt","w")
arr=list(map(int,inp.readline().split()))
adj=adjacency_list(arr[0],arr[1])
src=list(map(int,inp.readline().split()))
src1=src[0]
src2=src[1]


heap1=[(0,src1)]
heap2=[(0,src2)]
heapq.heapify(heap1)
heapq.heapify(heap2)
distance1=[999999]*(arr[0]+1)
visited1=[False]*(arr[0]+1)
distance1[src1]=0
visited1[src1]=True
while len(heap1)>0:
    src1=heapq.heappop(heap1)[1]
    distance1=dijkstra(adj,src1,heap1,distance1,visited1)

distance2=[999999]*(arr[0]+1)
visited2=[False]*(arr[0]+1)
distance2[src2]=0
visited2[src2]=True
while len(heap2)>0:
    src2=heapq.heappop(heap2)[1]
    distance2=dijkstra(adj,src2,heap2,distance2,visited2)

#print(distance1,distance2)


time=999999
node=0
for i in range(1,len(distance1)):
    if max(distance1[i],distance2[i])<time:
        time=max(distance1[i],distance2[i])
        node=i
if time==999999:
    out.write("Impossible")
else:
    out.write("Time "+ str(time)+"\n")
    out.write("Node "+ str(node))
inp.close()
out.close()
```
