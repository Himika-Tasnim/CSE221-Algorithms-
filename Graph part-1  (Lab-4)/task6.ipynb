```python
def count_diamond(adj_list,source,vertex,lst):
  visited=[False]*(vertex+1)
  queue=[source]
  visited[source]=True
  count=0 

  while len(queue)!=0:
    for i in (adj_list[queue[0]]):
      if visited[i]==False:
        queue.append(i)
        visited[i]=True
    if lst[queue[0]]=="D":
      count+=1
    queue.pop(0)

  return count

def graph_create(row,col):
  lst2D=[]
  lst=[None]
  for i in range(row):
    temp=str(inp.readline())
    temp=temp[:-1]
    x=[]
    for j in temp:
      x.append(j)
      lst.append(j)
    lst2D.append(x)
  

  adj=[]
  vertex=row*col
  for j in range(vertex+1):
    adj.append([])

  for i in range(len(lst2D)):
    temp=lst2D[i]
    for k in range(len(temp)):

      #connect with left
      if 0<=k-1:
        x=col*i+k
        y=col*i+k+1
        if temp[k-1]!="#":
          adj[y].append(x)

      #connect with right
      if k+1<len(temp):
        x=col*i+k+2
        y=col*i+k+1
        if temp[k+1]!="#":
          adj[y].append(x)

    #connect with up
    if 0<=i-1:
      for k in range(len(temp)):
        x=col*(i-1)+k+1
        y=col*i+k+1
        if lst2D[i-1][k]!="#":
          adj[y].append(x)

    #connect with down
    if i+1<len(lst2D):
      for k in range(len(temp)):
        x=col*(i+1)+k+1
        y=col*i+k+1
        if lst2D[i+1][k]!="#":
          adj[y].append(x)
 
  vertex=row*col
  c=str(count_diamond(adj,1,vertex,lst))
  return c

  

inp=open("input6.txt","r")
out=open("output6.txt","w")
array=list(map(int,inp.readline().split()))
count=graph_create(array[0],array[1])
out.write(count)
inp.close()
out.close()

```
