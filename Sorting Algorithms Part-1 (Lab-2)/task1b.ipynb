```python

def merge(lst1,lst2):
  f=len(lst1)+len(lst2)
  lst1.append(10000000)
  lst2.append(10000000)
  final=[]
  i=0
  j=0
  for x in range(f):
    if lst1[i]<lst2[j]:
      final.append(lst1[i])
      i+=1
    else:
      final.append(lst2[j])
      j+=1

  return final

def mergeSort(arr):
  if len(arr) <= 1:
    return arr
  else:
    mid = len(arr)//2
    a1 = mergeSort(arr[:mid])  # write the parameter
    a2 = mergeSort(arr[mid:])  # write the parameter
    return merge(a1, a2)


inp=open("input1b.txt","r")
out=open("output1b.txt","w")
x=inp.readline().split()
sum=int(x[1])
arr=list(map(int,inp.readline().split()))
sorted=mergeSort(arr)
start=0
end=len(sorted)-1
flag=False

while start<end:
  if sorted[start]+sorted[end]==sum:
    flag=True
    break
  elif sorted[start]+sorted[end]>sum:
    end-=1
  else:
    start+=1

#print(sorted[start],sorted[end])

if flag==True:
   for i in range(len(arr)):
    if arr[i]==sorted[start]:
      start=i+1
      break
   for i in range(len(arr)):
    if arr[i]==sorted[end]:
      end=i+1
      break
    
   y=f"{start} {end}"

else:
  y="IMPOSSIBLE"
    
out.write(y)

inp.close()
out.close()
```
