```python
def adjacency_list(vertex,edge):
  adj_list=[]
  for i in range(vertex+1):
    adj_list.append([])
  for j in range(edge):
    temp=list(map(int,inp.readline().split()))
    row=temp[0]
    col=temp[1]
    adj_list[row].append(col)
    adj_list[col].append(row)
  return adj_list



def find_shortest_path(lst,source,vertex,destination):
  visited=[False]*(vertex+1)
  queue=[source]
  visited[source]=True
  parent_list=[0]*(vertex+1)

  while len(queue)!=0:
    top=queue[0]
    parent_list[source]=source
    for i in (lst[top]):
      if visited[i]==False:
        queue.append(i)
        visited[i]=True
        parent_list[i]=top
      if i==destination:
        break
    queue.pop(0)

  
  temp=[]
  k=destination
  while k!=source:
    temp.append(k)
    k=parent_list[k]
  temp.append(source)
  temp.reverse()
  return temp


inp=open("input5.txt","r")
out=open("output5.txt","w")

array=list(map(int,inp.readline().split()))

adj_list=adjacency_list(array[0],array[1])
path=find_shortest_path(adj_list,1,array[0],array[2])
time="Time: "+str(len(path)-1)+"\n"
out.write(time)
out.write("Shortest Path: ")
for i in path:
  out.write(str(i)+" ")
inp.close()
out.close()

```
